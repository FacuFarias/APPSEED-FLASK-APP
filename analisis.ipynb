```python
def extraer(nombre, argumento):
    return nombre.replace(argumento, '')

# Aplicar la función a cada fila del DataFrame
df['ProductName'] = df.apply(lambda row: extraer(row['ProductName'], (row['ProductBrand'])+' '), axis=1)
```

```python
import mysql.connector
from sqlalchemy import create_engine, Column, Integer, String,DateTime,TIMESTAMP
from sqlalchemy.ext.declarative import declarative_base

# Definir el modelo de la tabla
Base = declarative_base()


#Comento esta tabla por que ya esta creada
'''class Productos(Base):
    __tablename__ = 'productos'
    id = Column(Integer, primary_key=True)
    ProductName = Column(String(255))
    ColorID = Column(Integer())
    MarcaID = Column(Integer())
    GeneroID = Column(Integer())
    Price=Column(Integer())
    Stock = Column(Integer())
    Description = Column(String(2500))'''

class Productos(Base):
    __tablename__ = 'colores'
    ColorID = Column(Integer, primary_key=True)
    ColorName = Column(String(50))

class Genero(Base):
    __tablename__ = 'genero'
    GeneroID = Column(Integer, primary_key=True)
    Genero = Column(String(50))

class Marcas(Base):
    __tablename__ = 'marcas'
    MarcaID = Column(Integer, primary_key=True)
    Marca = Column(String(50))

class Vendedores(Base):
    __tablename__ = 'vendedores'
    VendedorID = Column(Integer, primary_key=True)
    Nombre = Column(String(100))
    Sexo = Column(String(2))
    Edad = Column(Integer())

class Ventas(Base):
    __tablename__ = 'ventas'
    Transaccion = Column(Integer, primary_key=True)
    Fecha = Column(DateTime)
    ProdID = Column(Integer())
    vendedorID = Column(Integer())
```

```python
engine = create_engine('mysql+mysqlconnector://admin:1234@:3306/db_stockvision')
Base.metadata.create_all(engine)
```

```python
#Rellenamos las tablas
df_vendedores.to_sql('vendedores', con=engine, if_exists='replace', index=False)
df_ventas.to_sql('ventas', con=engine, if_exists='replace', index=False)
df_marcas.to_sql('marcas', con=engine, if_exists='replace', index=False)
df_genero.to_sql('genero', con=engine, if_exists='replace', index=False)
df_colores.to_sql('colores', con=engine, if_exists='replace', index=False)
```
